```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

users = {
    "Meena": [5, 4, 3, 2],
    "Kavya": [2, 3, 5, 4]
}


dresses = {
    "Party Frock": [5, 3, 5, 2],
    "Casual Wear": [3, 5, 2, 4],
    "Designer Gown": [4, 2, 5, 1],
    "Summer Dress": [4, 5, 3, 3],
    "Budget Dress": [2, 3, 2, 5]
}


U = np.array(list(users.values()))
D = np.array(list(dresses.values()))

sim = cosine_similarity(U, D)


for i, user in enumerate(users):

    scores = sorted(
        zip(dresses, sim[i]),
        key=lambda x: x[1],
        reverse=True
    )[:2]

    print(user, "->",
          [(d, round(s, 3)) for d, s in scores])
```
