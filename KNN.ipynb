```python
from sklearn.neighbors import KNeighborsClassifier

X = [[25,20],[63,120],[33,75],[42,100]]
y = [0,1,0,1]

model = KNeighborsClassifier(n_neighbors=3)

model.fit(X,y)

result = model.predict([[50,140]])

print(result)
```
